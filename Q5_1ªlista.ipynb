```python
import numpy as np
import matplotlib.pyplot as plt

# Definir a matriz da imagem
imagem = np.array([
    [0, 1, 1, 2, 1],
    [0, 0, 0, 2, 0],
    [0, 2, 2, 0, 0],
    [8, 8, 8, 8, 1]
])

# Calcular o histograma
histograma = np.histogram(imagem, bins=range(10))[0]

# Calcular o histograma acumulado
histograma_acumulado = np.cumsum(histograma)

# Imprimir os resultados
print("Histograma:")
print(histograma)
print("\nHistograma Acumulado:")
print(histograma_acumulado)

# Plotar o histograma e o histograma acumulado lado a lado
plt.figure(figsize=(15, 5))

# Plotar o histograma
plt.subplot(1, 2, 1)
plt.bar(range(len(histograma)), histograma, color='blue', alpha=0.7)
plt.title('Histograma')
plt.xlabel('Intensidade de Cor')
plt.ylabel('Contagem de Pixels')
plt.xticks(range(len(histograma)))
plt.grid(True)

# Plotar o histograma acumulado
plt.subplot(1, 2, 2)
plt.bar(range(len(histograma_acumulado)), histograma_acumulado, color='green', alpha=0.7)
plt.title('Histograma Acumulado')
plt.xlabel('Intensidade de Cor')
plt.ylabel('Contagem de Pixels Acumulada')
plt.xticks(range(len(histograma_acumulado)))
plt.grid(True)

plt.tight_layout()
plt.show()
```
