```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Definir a matriz da imagem
imagem = np.array([
    [0, 1, 1, 2, 1],
    [0, 0, 0, 2, 0],
    [0, 2, 2, 0, 0],
    [8, 8, 8, 8, 1]
], dtype=np.uint8)  # Certifique-se de usar o tipo de dados uint8

# Criar um histograma da imagem original
hist_original, bins = np.histogram(imagem.flatten(), bins=range(10))

# Calcula a função de distribuição acumulada (CDF)
cdf = np.cumsum(hist_original)

# Normalizar CDF para estar no intervalo [0,255]
cdf_normalized = ((cdf - cdf.min()) * 255) / (cdf.max() - cdf.min())

# Aplicar a equalização por ordenação na imagem
imagem_equalizada = cdf_normalized[imagem]

# Criar um histograma da imagem equalizada
hist_equalizado, _ = np.histogram(imagem_equalizada.flatten(), bins=range(10))

# Plotar os histogramas
h_eq = cv2.equalizeHist(img)

# Configurar a figura e os subplots
plt.figure(figsize=(12, 6))

# Plotar o histograma original
plt.subplot(1, 2, 2)
plt.title("Histograma Original")
plt.xlabel("Intensidade")
plt.ylabel("Qtde de Pixels")
plt.hist(img.ravel(), 256, [0, 256])
plt.xlim([0, 256])


# Plotar o histograma equalizado
plt.subplot(1, 2, 1)
plt.title("Histograma Equalizado")
plt.xlabel("Intensidade")
plt.ylabel("Qtde de Pixels")
plt.hist(h_eq.ravel(), 256, [0, 256])
plt.xlim([0, 256])


# Exibir os subplots
plt.tight_layout()
plt.show()

cv2.waitKey(0)

# Plotar as imagens antes e depois da equalização
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(imagem_equalizada, cmap='gray')
plt.title('Imagem Equalizada')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagem, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')


plt.tight_layout()
plt.show()

```
