```python
import numpy as np
import matplotlib.pyplot as plt

# Definir as trs imagens
imagem1 = np.array([
    [0, 1, 1, 2, 1],
    [0, 0, 0, 2, 0],
    [0, 2, 2, 0, 0],
    [8, 8, 8, 8, 1]
])

imagem2 = np.array([
    [2, 2, 4, 4],
    [3, 2, 4, 4],
    [3, 1, 2, 3],
    [1, 1, 1, 3],
    [4, 4, 4, 4]
])

imagem_final = np.array([
    [2, 3, 3, 4, 3],
    [2, 2, 2, 4, 2],
    [2, 4, 4, 2, 2],
    [4, 4, 4, 4, 3]
])

# Calcular histogramas acumulados
hist_acumulado1 = np.cumsum(np.histogram(imagem1, bins=range(10))[0])
hist_acumulado2 = np.cumsum(np.histogram(imagem2, bins=range(10))[0])
hist_acumulado_final = np.cumsum(np.histogram(imagem_final, bins=range(10))[0])

# Plotar histogramas acumulados e representao grfica
plt.figure(figsize=(15, 10))

# Histograma acumulado e representao grfica da imagem 1
plt.subplot(2, 3, 1)
plt.plot(hist_acumulado1, color='blue')
plt.title('Histograma Acumulado - Imagem da Q5')
plt.xlabel('Intensidade de Cor')
plt.ylabel('Nmero de Pixels')
plt.grid(True)

# Histograma da imagem 1
plt.subplot(2, 3, 4)
plt.hist(imagem1.ravel(), bins=range(10), color='blue')
plt.title('Histograma - Imagem da Q5')
plt.xlabel('Intensidade de Cor')
plt.ylabel('Nmero de Pixels')

# Histograma acumulado e representao grfica da imagem 2
plt.subplot(2, 3, 2)
plt.plot(hist_acumulado2, color='green')
plt.title('Histograma Acumulado - Imagem da Q8')
plt.xlabel('Intensidade de Cor')
plt.ylabel('Nmero de Pixels')
plt.grid(True)

# Histograma da imagem 2
plt.subplot(2, 3, 5)
plt.hist(imagem2.ravel(), bins=range(10), color='green')
plt.title('Histograma - Imagem da Q8')
plt.xlabel('Intensidade de Cor')
plt.ylabel('Nmero de Pixels')

# Histograma acumulado e representao grfica da imagem final
plt.subplot(2, 3, 3)
plt.plot(hist_acumulado_final, color='red')
plt.title('Histograma Acumulado - Imagem Casada')
plt.xlabel('Intensidade de Cor')
plt.ylabel('Nmero de Pixels')
plt.grid(True)

# Histograma da imagem final
plt.subplot(2, 3, 6)
plt.hist(imagem_final.ravel(), bins=range(10), color='red')
plt.title('Histograma - Imagem Casada')
plt.xlabel('Intensidade de Cor')
plt.ylabel('Nmero de Pixels')

plt.tight_layout()
plt.show()

# Plotar as imagens
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(imagem1, cmap='gray')
plt.title('Imagem da Q5')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(imagem2, cmap='gray')
plt.title('Imagem da Q8')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(imagem_final, cmap='gray')
plt.title('Imagem do casamento')
plt.axis('off')

plt.tight_layout()
plt.show()
```
