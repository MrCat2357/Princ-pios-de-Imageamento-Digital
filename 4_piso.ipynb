```python
from IPython.display import display, Image
import ipywidgets as widgets
from IPython.display import clear_output

# Caminhos para as imagens
image_paths = {
    'central': '/content/drive/MyDrive/raq/4piso.png',
    'image1': '/content/drive/MyDrive/raq/im1.png',
    'image2': '/content/drive/MyDrive/raq/im2.png',
    'image3': '/content/drive/MyDrive/raq/im3.png',
    'image4': '/content/drive/MyDrive/raq/im4.png',
    'image5': '/content/drive/MyDrive/raq/im5.png',
    'image6': '/content/drive/MyDrive/raq/im6.png',
    'image7': '/content/drive/MyDrive/raq/im7.png',
    'image8': '/content/drive/MyDrive/raq/im8.png',
    'image9': '/content/drive/MyDrive/raq/im9.png',
    'image10': '/content/drive/MyDrive/raq/im10.png',
    'image11': '/content/drive/MyDrive/raq/im11.png',
    'image12': '/content/drive/MyDrive/raq/im12.png',
    'image13': '/content/drive/MyDrive/raq/im13.png',
    'image14': '/content/drive/MyDrive/raq/im14.png',
    'image15': '/content/drive/MyDrive/raq/im15.png',
    'image16': '/content/drive/MyDrive/raq/im16.png',
    'image17': '/content/drive/MyDrive/raq/im17.png',
    'image18': '/content/drive/MyDrive/raq/im18.png',
    'image19': '/content/drive/MyDrive/raq/im19.png',
    'image20': '/content/drive/MyDrive/raq/im20.png',
}

# Função para atualizar a imagem exibida
def show_image(image_path):
    with image_output:
        clear_output(wait=True)
        display(Image(filename=image_path))

# Botões para alternar imagens
buttons = [widgets.Button(description=str(i)) for i in range(1, 21)]

# Vincular botões à função de exibição de imagem
for i, button in enumerate(buttons):
    button.on_click(lambda x, i=i: show_image(image_paths[f'image{i+1}']))

# Layout dos botões ao redor da imagem central
button_box = widgets.VBox([
    widgets.HBox(buttons[:10]),
    widgets.Image(value=open(image_paths['central'], 'rb').read(), format='png'),
    widgets.HBox(buttons[10:])
])

# Output widget para a imagem selecionada
image_output = widgets.Output()

# Layout final
final_layout = widgets.VBox([button_box, image_output])

# Exibir layout
display(final_layout)
```
